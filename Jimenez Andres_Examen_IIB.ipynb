```python
import pandas as pd

file_path = "/content/arxiv_dataset/arxiv-metadata-oai-snapshot.json"

sample_lines = int(2150000 * 0.01)

# Usar chunks para cargar sólo una parte
df_iter = pd.read_json(file_path, lines=True, chunksize=sample_lines)
df_chunk = next(df_iter)

# Usar 'abstract' en lugar de 'summary'
df = df_chunk[["id", "title", "abstract", "authors", "categories"]].dropna().reset_index(drop=True)

df.head()

```

## 1. Preprocesamiento

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Descargar recursos necesarios
nltk.download("punkt")
nltk.download("stopwords")

# Lista de stopwords y puntuación
stop_words = set(stopwords.words("english"))
punctuation = set(string.punctuation)
```

```python
def preprocess_text(text):
    text = text.lower()  # Minúsculas
    tokens = word_tokenize(text)  # Tokenización
    tokens = [word for word in tokens if word not in stop_words and word not in punctuation]
    return " ".join(tokens)
```

```python
import nltk
nltk.download('punkt_tab')

df["text"] = df["title"] + " " + df["abstract"] + " " + df["authors"] + " " + df["categories"]

# Aplicar preprocesamiento
df["processed_text"] = df["text"].apply(preprocess_text)

# Mostrar los primeros resultados
df[["id", "processed_text"]].head()
```

## 2. Indexacion y Embeddings


```python
# Tf-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df["processed_text"])
```

```python
#BM25
from rank_bm25 import BM25Okapi

tokenized_corpus = [doc.split() for doc in df["processed_text"]]
bm25 = BM25Okapi(tokenized_corpus)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Cargar modelo (pequeño y eficiente)
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Generar vectores (embeddings)
embeddings = embedding_model.encode(df["processed_text"], show_progress_bar=True)

# Crear índice FAISS
faiss_index = faiss.IndexFlatL2(embeddings.shape[1])
faiss_index.add(np.array(embeddings))
```

## 3. Funciones de Recuperacion

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from rank_bm25 import BM25Okapi
import faiss

def normalize_embeddings(embeddings):
    norms = np.linalg.norm(embeddings, axis=1, keepdims=True)
    return embeddings / norms

def preprocess_query(query):
    return preprocess_text(query)

def search_tfidf(query, top_k=10):
    query_proc = preprocess_query(query)
    vec_query = tfidf_vectorizer.transform([query_proc])
    scores = cosine_similarity(tfidf_matrix, vec_query).ravel()
    top_indices = np.argsort(scores)[::-1][:top_k]
    results = df.iloc[top_indices][["id", "title", "abstract"]].copy()
    results["score"] = scores[top_indices]
    return results

def search_bm25(query, top_k=10):
    query_tokens = preprocess_query(query).split()
    scores = bm25.get_scores(query_tokens)
    top_indices = np.argsort(scores)[::-1][:top_k]
    results = df.iloc[top_indices][["id", "title", "abstract"]].copy()
    results["score"] = scores[top_indices]
    return results

def search_faiss(query, top_k=10):
    query_proc = preprocess_query(query)
    query_vec = embedding_model.encode([query_proc])
    query_vec_norm = query_vec / np.linalg.norm(query_vec, axis=1, keepdims=True)
    query_vec_norm = query_vec_norm.astype('float32')
    D, I = faiss_index_cosine.search(query_vec_norm, top_k)
    results = df.iloc[I[0]][["id", "title", "abstract"]].copy()
    results["score"] = D[0]  # Add the scores to the results DataFrame
    return results
```

```python
queries = [
    "diphoton production cross sections",
    "quantum chromodynamics",
    "higgs boson decay",
    "machine learning for particle physics",
    "top quark production"
]

# Crear el archivo queries.txt
with open("queries.txt", "w") as f:
    for q in queries:
        f.write(q + "\n")
```

```python
def print_results(queries_file="queries.txt", top_k=10):
    with open(queries_file, "r") as f:
        queries = [line.strip() for line in f.readlines() if line.strip()]

    for query in queries:
        print(f"\n{'='*100}\nQuery: {query}\n{'='*100}")

        for method, search_func in [
            ("TF-IDF", search_tfidf),
            ("BM25", search_bm25),
            ("FAISS", search_faiss)
        ]:
            print(f"\n--- {method} Results ---")
            results = search_func(query, top_k=top_k)
            for _, row in results.iterrows():
                snippet = row["abstract"][:300].replace("\n", " ")
                print(f"Score: {row['score']:.4f} | ID: {row['id']}\nTitle: {row['title']}\nAbstract: {snippet}...\n")
```

## 4. RAG

```python
def generar_contexto_faiss(query, top_k=3):
    # Codificar el embedding de la consulta
    query_proc = preprocess_query(query)
    query_vec = embedding_model.encode([query_proc])
    query_vec_norm = query_vec / np.linalg.norm(query_vec, axis=1, keepdims=True)
    query_vec_norm = query_vec_norm.astype('float32')

    # Buscar en el índice FAISS
    D, I = faiss_index_cosine.search(query_vec_norm, top_k)

    # Extraer los documentos relevantes
    textos_topk = []
    for i in I[0]:
        row = df.iloc[i]
        texto = f"Title: {row['title']}\nAbstract: {row['abstract']}"
        textos_topk.append(texto)

    # Concatenar con separadores
    contexto = "\n\n-----\n\n".join(textos_topk)

    # Mostrar el contexto preparado
    print("Contexto para el LLM:")
    print(contexto[:1500] + "\n...") if len(contexto) > 1500 else print(contexto)

    return contexto
```

## 5. Evaluacion

```python
def evaluar_queries(archivo="queries.txt", top_k=10):
    with open(archivo, "r") as f:
        queries = [line.strip() for line in f if line.strip()]

    for query in queries:
        print(f"\n{'='*120}\n🧪 Evaluación para la consulta: '{query}'\n{'='*120}")

        # Obtener resultados
        r_tfidf = search_tfidf(query, top_k)
        r_bm25 = search_bm25(query, top_k)
        r_faiss = search_faiss(query, top_k)

        # IDs
        ids_tfidf = list(r_tfidf["id"])
        ids_bm25 = list(r_bm25["id"])
        ids_faiss = list(r_faiss["id"])

        set_tfidf, set_bm25, set_faiss = set(ids_tfidf), set(ids_bm25), set(ids_faiss)

        # Comparación de intersecciones
        comunes_tf_bm = set_tfidf & set_bm25
        comunes_tf_fa = set_tfidf & set_faiss
        comunes_bm_fa = set_bm25 & set_faiss
        comunes_todos = set_tfidf & set_bm25 & set_faiss

        print("🔁 Documentos en común:")
        print(f"TF-IDF ∩ BM25: {len(comunes_tf_bm)} documentos")
        print(f"TF-IDF ∩ FAISS: {len(comunes_tf_fa)} documentos")
        print(f"BM25 ∩ FAISS: {len(comunes_bm_fa)} documentos")
        print(f"TF-IDF ∩ BM25 ∩ FAISS: {len(comunes_todos)} documentos")

        # Comparar orden
        print("\n📊 Ordenamiento de IDs (top 5):")
        print("TF-IDF:", ids_tfidf[:5])
        print("BM25  :", ids_bm25[:5])
        print("FAISS :", ids_faiss[:5])

        # Medir similitud entre rankings
        def contar_coincidencias(lista1, lista2):
            return sum(1 for doc in lista1 if doc in lista2)

        print("\n🔢 Similitud entre rankings (top-10):")
        print(f"TF-IDF vs BM25 : {contar_coincidencias(ids_tfidf, ids_bm25)} coincidencias")
        print(f"TF-IDF vs FAISS: {contar_coincidencias(ids_tfidf, ids_faiss)} coincidencias")
        print(f"BM25 vs FAISS  : {contar_coincidencias(ids_bm25, ids_faiss)} coincidencias")

        # Tabla comparativa
        print("\n📋 Tabla comparativa (top-5):")
        print("{:<6} {:<15} {:<15} {:<15}".format("Rank", "TF-IDF", "BM25", "FAISS"))
        for i in range(5):
            print("{:<6} {:<15} {:<15} {:<15}".format(
                i+1,
                ids_tfidf[i] if i < len(ids_tfidf) else "-",
                ids_bm25[i] if i < len(ids_bm25) else "-",
                ids_faiss[i] if i < len(ids_faiss) else "-"
            ))
```
